```python
import pandas as pd
diabetes=pd.read_csv("diabetes.csv")
```

```python
#dropping pregnancies & outcome before converting zeros to NaN
#pregnancy column dropped because filling in with median value doesn't make sense 
diabetes_num= diabetes.drop(['Pregnancies', 'Outcome'], axis=1)
```

```python
import numpy as np
diabetes_num.replace(0, np.nan, inplace=True)
```

```python
#replacing NaN values with the median of each column
diabetes_num = diabetes_num.fillna(diabetes.median())
```

```python
diabetes_num.describe()
```

```python
#standardizing the data
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
diabetes_tr= scaler.fit_transform(diabetes_num)
```

```python
diabetes_tr_df= pd.DataFrame(diabetes_tr, columns= diabetes_num.columns, index= diabetes_num.index)
diabetes_tr_df.head()
```

```python
X=diabetes_num
X=X.values
y=diabetes["Outcome"]
y=y.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=12345)
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
train_preds=knn_model.predict(X_train)
mse=mean_squared_error(y_train,train_preds)
rmse=sqrt(mse)
print("Error rate on training data :",rmse)
```

```python
test_preds=knn_model.predict(X_test)
mse=mean_squared_error(y_test,test_preds)
rmse=sqrt(mse)
print("Error rate on testing data :",rmse)
```
